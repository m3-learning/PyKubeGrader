```python
from pykubegrader.widgets.true_false import TFQuestion, TrueFalse_style
```

```python
class Question1(TFQuestion):

    def __init__(
        self,
        title="Respond with True or False",
        style=TrueFalse_style,
        question_number=2,
        keys=["MC1", "MC2", "MC3", "MC4"],
        descriptions=[
            "Which of the following stores key:value pairs?",
            "The following condition returns to the next iteration of the loop",
            "Which operator is used for exponentiation in Python?",
            "Which method is used to add an element to the end of a list in Python?",
        ],
        points=2,
    ):
        super().__init__(
            title=title,
            style=style,
            question_number=question_number,
            keys=keys,
            options=[["True", "False"] for _ in range(len(keys))],
            descriptions=descriptions,
            points=points,
        )
```

```python
import pykubegrader.initialize
import panel as pn

pn.extension()

Question1().show()
```

```python
import panel as pn

# Define Drexel theme colors including the `--jp-brand-color0`
drexel_colors = """
:root {
    --jp-brand-color0: #07294D; /* Drexel Blue */
    --jp-brand-color1: #FFC600; /* Drexel Gold */
    --jp-layout-color0: #F8F8F8; /* Background Color */
    --jp-layout-color1: #FFFFFF; /* Surface Color */
    --jp-brand-text-color: #FFFFFF; /* Text on Brand Color */
    --jp-layout-text-color: #07294D; /* Text on Background */
}

/* Apply custom styles to ensure alignment with Drexel colors */
body {
    background-color: var(--jp-layout-color0);
    color: var(--jp-layout-text-color);
}

.bk.card {
    background-color: var(--jp-brand-color1);
    color: var(--jp-layout-text-color);
    border-radius: 8px;
    padding: 10px;
}

.bk-btn {
    background-color: var(--jp-brand-color0);
    color: var(--jp-brand-text-color);
    border: none;
    border-radius: 4px;
    cursor: pointer;
}

.bk-btn:hover {
    background-color: var(--jp-brand-color1);
    color: var(--jp-brand-color0);
}

/* Styling for RadioBoxGroup */
.bk-radio-group .bk-input {
    appearance: none;
    -webkit-appearance: none;
    background-color: white;
    border: 2px solid var(--jp-brand-color0);
    border-radius: 50%;
    width: 16px;
    height: 16px;
    margin: 2px 6px 2px 0;
    display: inline-block;
    vertical-align: middle;
    cursor: pointer;
}
.bk-radio-group .bk-input:checked {
    background-color: var(--jp-brand-color0);
    border-color: var(--jp-brand-color1);
}
.bk-radio-group .bk-input:hover {
    background-color: var(--jp-brand-color1);
}
.bk-radio-group label {
    color: var(--jp-brand-color0);
    font-weight: bold;
    margin-left: 8px;
    vertical-align: middle;
    cursor: pointer;
}
"""

# Pass the custom CSS to Panel
pn.extension(design="material", global_css=[drexel_colors])


# Create sample layout
def test_layout():
    card = pn.Card(
        "This card uses `--jp-brand-color1` (Drexel Gold)", title="Drexel Colors"
    )
    button = pn.widgets.Button(name="Click Me", button_type="primary")
    radio_buttons = pn.widgets.RadioBoxGroup(
        name="Choose an option",
        options=["Option 1", "Option 2", "Option 3"],
        sizing_mode="stretch_width",
    )
    layout = pn.Column(
        pn.pane.Markdown("# Drexel Colors with JupyterLab Variables"),
        card,
        button,
        radio_buttons,
        sizing_mode="stretch_width",
    )
    return layout


# Serve the layout
test_layout().servable()
```
